#### Utilidad de NOTEBOOK

Esto nos permitiria tener un notebook prinicipal, y que al momento de ejecutar , llame e ejecute  
a otros notebooks y deise√±ar una aplicacion modular

```python
dbutils.notebook.help()
```

```python
#dbutils.notebook.run("ruta", tiempo de espera en segundos,para pasar parametros al notebook hijo)
#Este notebook ejecuta a otro notebook hijo
dbutils.notebook.run("./Utilities-child",25)
```

#### Utilidad de WIDGETS

Esto nos permitira pasar parametros a los notebooks

```python
dbutils.widgets.help()
```

```python
# ESTAS CELDAS NO EJECUTAR PORQUE ESTO ESTA EXPLICADO EN EL NOTEBOOK HIJO
# AL MOMNETO DE LLAMARLO LO VAMOS A PASAR ESOS PARAMETROS

#dbutils.widgets.text(nombre_variable,valor por defecto,etiqueta widgets)
dbutils.widgets.text("v_msg","","Parametro entrada")
```

```python
# PARA CAPTURAR ESE VALOR
v_nombre=dbutils.widgets.get("v_msg")
```

```python
#VAMOS A EJECUTAR EL NOTEBOOK HIJO QUE ESTA COMPUESTO POR PARAMETROS Y LOS VAMOS A PASAR
#DESDE ESTE NOTEBOOK PARA QUE PUEDA SER EJECUTADO
dbutils.notebook.run("./Utilities-child",25,{"v_msg":"Hola mundo"})
```
